```python

df = pd.read_csv('Child_2024.csv')
df['Ben Created'] = pd.to_datetime(df['Ben Created'], format='%d-%b-%Y')

df_sorted = df.sort_values(by='Ben Created', ascending=False)
print(df_sorted)

df_sorted.to_csv('Child_2024_sorted.csv', index=False)
```

```python

df_sorted = pd.read_csv('Child_2024_sorted.csv')

df_deduped = df_sorted.drop_duplicates(subset='Beneficiary Id', keep='first')
df_deduped.to_csv('Child_2024_latest.csv', index=False)

```

```python

df_child = pd.read_csv('Child_2024_latest.csv')
df_clean = pd.read_csv('CClean.csv')

df_child['Beneficiary Id'] = df_child['Beneficiary Id'].astype(str).str.strip()
df_clean['Beneficiary Id'] = df_clean['Beneficiary Id'].astype(str).str.strip()

common_cols = [col for col in df_child.columns if col in df_clean.columns and col != 'Beneficiary Id']
df_child_reduced = df_child.drop(columns=common_cols)

df_merged = pd.merge(df_child_reduced, df_clean, on='Beneficiary Id', how='inner')

df_merged.to_csv('Final_Clean.csv', index=False)


```
